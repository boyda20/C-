```python
!pip install --quiet PyMuPDF

import fitz  # PyMuPDF

def extract_text_with_bounding_boxes(pdf_path):
    doc = fitz.open(pdf_path)
    extracted_data = []

    for page_num in range(doc.page_count):
        page = doc[page_num]

        # Extract text blocks which often include bounding box information
        # 'textpage.extractBLOCKS()' is useful for structured text extraction
        text_blocks = page.get_text("blocks") # 'blocks' format returns (x0, y0, x1, y1, text, block_no, block_type)

        page_data = []
        for block in text_blocks:
            x0, y0, x1, y1, text, block_no, block_type = block
            bbox = (x0, y0, x1, y1)
            page_data.append({
                'text': text.strip(),
                'bbox': bbox
            })
        extracted_data.append({
            'page_number': page_num + 1,
            'content': page_data
        })

    doc.close()
    return extracted_data

# Path to the downloaded PDF
pdf_file = output_filename

# Extract text and bounding boxes
loan_worksheet_data = extract_text_with_bounding_boxes(pdf_file)

# Display the extracted data for the first page as an example
if loan_worksheet_data:
    print(f"Extracted data for page 1:\n{loan_worksheet_data[0]['content']}")
else:
    print("No data extracted.")
```
