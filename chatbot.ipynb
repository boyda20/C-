```python
from google.colab import userdata
import os

# Assuming you have stored your API key as a secret named 'GOOGLE_API_KEY'
# in Colab's Secret Manager.
# If your secret has a different name, replace 'GOOGLE_API_KEY' below.
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
os.environ['GOOGLE_API_KEY'] = GOOGLE_API_KEY
```

```python
import google.generativeai as genai
llm = genai.GenerativeModel(
    model_name="gemini-2.5-flash"
)
```

```python
def simple_chatbot():
    """
    A simple interactive chatbot using Gemini.
    """
    print("🦙 Simple Gemini Chatbot 🦙")
    print("Type 'exit' to end the conversation")
    print("-" * 50)

    # Initialize chat history for the genai model
    chat_session = llm.start_chat(history=[])

    while True:
        # Get user input
        user_input = input("\nYou: ")

        # Check for exit command
        if user_input.lower() in ["exit", "quit", "bye"]:
            print("\nChatbot: Goodbye! Have a great day!")
            break

        try:
            # Get response from Gemini
            response = chat_session.send_message(user_input)

            # Print the response
            print(f"\nChatbot: {response.text}")

        except Exception as e:
            print(f"\nError: {e}")
            print("Please try again or check your API key")
```

```python
# Run the chatbot
if __name__ == "__main__":
  simple_chatbot()
```
